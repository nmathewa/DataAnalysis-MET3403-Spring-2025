3. load the datasets and combine
4. correlation between reanlysis and observation

```python
#importing libraries 

import pandas as pd # for reading csv
import xarray as xr # for reading netcdf 
import metpy.calc as mcalc # metpy for calculating wind speed and direction
from metpy.units import units # import unit function utility from metpy
import matplotlib.pyplot as plt # importing matplotlib pyplot functions for plotting
```

```python

asos_data = pd.read_csv('MLB.csv',na_values='M')
asos_data['valid'] = pd.to_datetime(asos_data['valid']) # converting the time column to pandas datetime object

asos_data
```

```python
# load the ASOS 5 min data 

asos_data = asos_data.set_index('valid') # replaced the number based index with time 

asos_data
```

```python
# a quick preview 

asos_data['sknt'].plot()
```

```python
 # converting asos 5 min data in to hourly 

asos_without_station = asos_data.drop(['station','lon','lat'],axis=1) # dropping station since it's a string 

asos_hourly = asos_without_station.resample('1h').mean()

asos_hourly['sknt'].plot() # to plot 
```

```python
## 2. read the netcdf file 

era_reanalysis = xr.open_dataset('era_2024_dec.nc')# read the dataset

era_reanalysis
```

```python
# select nearest era5 point 

era_reanalysis_mlb = era_reanalysis.sel(longitude=-80.6,latitude=28.10,method='nearest')

era_reanalysis_mlb
```

```python
era_reanalysis_mlb_tabular = era_reanalysis.to_dataframe() # converted to a table 

era_reanalysis_mlb_tabular
```

```python
era_reanalysis_mlb_tabular.reset_index().set_index('valid_time').drop(['latitude','longitude','number','expver'],axis=1)  # to reset the index (multi-index ---> single index)
```

```python


final_frame = era_reanalysis_mlb_tabular.reset_index().set_index('valid_time').drop(['latitude','longitude','number','expver'],axis=1) # restructure the dataframe

final_frame
```

```python
# combine both datasets 
# calculate wind speed from u and v 

u_wind_era = final_frame['u10'].values * units('m/s') # assign units for u_wind

v_wind_era = final_frame['v10'].values * units('m/s') # assign units for v_wind

wind_speed_era = mcalc.wind_speed(u=u_wind_era,v=v_wind_era) # compute wind speed from metpy

wind_direction_era = mcalc.wind_direction(u=u_wind_era,v=v_wind_era) # compute wind direction from metpy 

```

```python
# adding the new wind speed and wind direction to the dataframe final_frame 

final_frame['wind_speed_era'] = wind_speed_era

final_frame['wind_direction_era'] = wind_direction_era

final_frame
```

```python
# Combine both dataset from asos and era5 

combined_dataset = pd.merge(asos_hourly,final_frame,left_index=True,right_index=True,how='inner') # merging dataset by index

combined_dataset = combined_dataset.drop(['tmpf','relh','p01m','u10','v10'],axis=1)

combined_dataset['wind_asos_mps'] = combined_dataset['sknt'] * 1.9 # converting knots to m/s

combined_dataset
```

```python
correlation_coeff = combined_dataset['wind_asos_mps'].corr(combined_dataset['wind_speed_era']) # correlation between wind speeds
```

To calculate the percentage of correlation 

$ percentage = r^{2} $

where $r^{2}$ is the correlation coefficient

```python
correlation_percentage = (correlation_coeff * correlation_coeff) * 100

print(f'{correlation_percentage} %')
```

```python
combined_dataset[['wind_asos_mps','wind_speed_era']].plot() # time series comparison
```

```python

fig,axis = plt.subplots() # create  a figure and axis object to control the plots

combined_dataset['wind_asos_mps'].hist(ax=axis,color='red',label='wind speed asos') # added

combined_dataset['wind_speed_era'].hist(ax=axis,alpha=0.7,label = 'wind speed era5')

axis.legend()

axis.set_xlabel(r'wind speed $ m s^{-1} $')
axis.set_title('Wind speed comparison between ASOS and ERA5 (KMLB)')

```
