```python
# importing all the libraries 
import xarray as xr # for reading the netcdf 
import pandas as pd # for reading tabular data / CSV file
import metpy.calc as mcalc #
```

```python

# read netcdf file using xarray 
hurricane_dataset = xr.open_dataset('HurricaneIAN.nc')


hurricane_dataset
```

```python
# print the time steps in the dataset
hurricane_dataset['time']
```

```python
# The dataset 2 hourly
# converting to daily 

hurricane_dataset_daily = hurricane_dataset.resample(time='D').mean() # mean over the 2 hpo

hurricane_dataset_daily
```

```python

# converting to monthly 

# converting to monthly (1st date of the month : MS)
hurricane_dataset_mons = hurricane_dataset.resample(time='MS').mean()

# converting to monthly (31st date of the month : MS)
hurricane_dataset_mone = hurricane_dataset.resample(time='ME').mean()


hurricane_dataset_mone
```

```python
# converting to 4 HOURLY 

# converting to 4 hourly from 2 hourly
hurricane_dataset_4hourly = hurricane_dataset.resample(time='4h').mean()


hurricane_dataset_4hourly
```

```python
# Group-by operation for seasonal 

# to calculate the seasonal mean 
seasonal_data = hurricane_dataset.groupby('time.season').mean()

seasonal_data
```

```python

# calculate the mean for every hour of a day (24 hour for a day) 
# here it is only 12 because 2 hourly timestep
diurnal_data = hurricane_dataset.groupby('time.hour').mean()

diurnal_data
```

```python
# How to save a daily mean temperature from 2 hourly data 

# 1. select the temperature from the dataset 
# 2. convert 2 hourly to daily 
# 3. convert to a table (csv) 
# 4. save the CSV 
```

```python
# Reading netcdf file 

hurricane_dataset = xr.open_dataset('HurricaneIAN.nc')

hurricane_dataset
```

```python

# selecting the temperature and convert to celsius
hurricane_temperature = ((hurricane_dataset['t'] - 273.15) * 1.8) + 32  # converting to F

# converting from 2 hourly to daily
temperature_daily = hurricane_temperature.resample(time='D').mean()
```

```python
# converting to table 

temperature_table = temperature_daily.to_dataframe()

temperature_table
```

```python

# save the csv file 
temperature_table.to_csv('temperature_daily_ian.csv')
```
